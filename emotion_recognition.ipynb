```python
from IPython.display import Video
from Katna.video import Video as KatnaVideo
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import urllib.request 
```

```python
data = scores.iloc[180]
print(data["video_url"])
display(Video(data["video_url"]))
videoPath = "./downloads/video_" + str(data["video_id"]) + ".mp4"
urllib.request.urlretrieve(data["video_url"], videoPath) 
vd = KatnaVideo()
imgs = vd.extract_frames_as_images(no_of_frames = 10, file_path=videoPath)

plt.figure(figsize=(20,10))
columns = 5
for i, img in enumerate(imgs):
    vd.save_frame_to_disk(
        img,
        file_path="./downloads/",
        file_name="video_" + str(data["video_id"]) + "_" + str(i),
        file_ext=".jpg",
    )
    plt.subplot(len(imgs) / columns + 1, columns, i + 1)
    plt.imshow(mpimg.imread('./downloads/video_'  + str(data["video_id"]) + "_" + str(i) + ".jpg"))
```
